```python
# Calculate the total number of transactions and total transaction amount for each state over the years
df_state_txn.groupby(['State', 'Year']).agg(total_transactions = ('Transactions', 'sum'), total_amount = ('Amount (INR)', 'sum')).reset_index()
```

```python
# Identify the top 5 states with the highest transaction volumes and the top 5 states with the lowest transaction volumes
df_state_txn.groupby(['State'])['Transactions'].sum().idxmax()
df_state_txn.groupby(['State'])['Transactions'].sum().nlargest(5)
```

```python
# Identify the most common transaction types in each state Year and quarter
top_idx = df_state_txn_split.groupby(['State', 'Year', 'Quarter'])['Transactions'].idxmax()

df_state_txn_split.loc[top_idx]
```

```python
# Identify the most common transaction types in each state Year
transactions =df_state_txn_split.groupby(['State', 'Year', 'Transaction Type'])['Transactions'].sum().reset_index()
rows = transactions.groupby(['State', 'Year'])['Transactions'].idxmax()
transactions.loc[rows]
```

```python
# Identify the most common transaction types in each state
transaction = df_state_txn_split.groupby(['State', 'Transaction Type'])['Transactions'].sum().reset_index()
rows = transaction.groupby(['State'])['Transactions'].idxmax()
transaction.loc[rows]
```

```python
# Determine the device brand with the highest number of registered users in each state
users = df_state_device.groupby(['State', 'Brand'])['Registered Users'].sum().reset_index()
rows = users.groupby(['State'])['Registered Users'].idxmax()
users.loc[rows]
```

```python
# Calculate the total number of transactions, total transaction amount, 
# and total registered users in each state by summing up the values from the district level data

df_district = df_district_txn.groupby(['State']).agg(district_transactions = ('Transactions', 'sum'), 
                                      district_transaction_amt = ('Amount (INR)', 'sum'),
                                      district_reg_users = ('Registered Users', 'sum')).reset_index()


# Calculate the total number of transactions, total transaction amount, 
# and total registered users in each state by summing up the values from the State level data

df_state = df_state_txn.groupby(['State']).agg(state_transactions = ('Transactions', 'sum'), 
                                      state_transaction_amt = ('Amount (INR)', 'sum'),
                                      state_reg_users = ('Registered Users', 'sum')).reset_index()


df_combined = df_state.merge(df_district, how = 'left', on = ['State'])
```

```python
# calculate the average transaction amount per user for each state

df_transactions = df_state_txn.groupby('State').agg(total_transactions = ('Transactions', 'sum'), 
                                  total_amount = ('Amount (INR)', 'sum')).reset_index()
df_transactions['avg_transaction_per_user'] = df_transactions['total_amount']/df_transactions['total_transactions']
df_transactions

df_transactions.nlargest(5, columns = 'avg_transaction_per_user')
df_transactions.nsmallest(5, columns = 'avg_transaction_per_user')
```
